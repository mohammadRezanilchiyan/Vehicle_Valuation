The Pearson correlation coefficient (r) is a value between -1 and 1 that represents the strength and direction of the linear relationship between two variables. 

The formula to calculate the Pearson correlation coefficient (r) is:

 <b>r = cov(X, Y) / (std(X) * std(Y))</b>
 

<li><b>r = +1:</b></li>A perfect positive linear relationship. As one variable increases, the other also increases linearly.
<li><b>r = 0:</b></li>No linear relationship. The variables are not correlated.
<li><b>r = -1:</b></li>A perfect negative linear relationship. As one variable increases, the other decreases linearly.

Dear student, looking at the distribution plot can be a helpful way to evaluate the fit of a model. Distribution plots, also known as density plots, display the distribution of data points along the range of a variable. They provide a visual representation of how the data is spread and can reveal important information about its shape, central tendency, and spread.

When assessing the fit of a model, particularly in the context of machine learning or statistical modeling, you would typically compare the distribution of the predicted values (output of the model) to the actual observed values (ground truth). Here are some common scenarios where distribution plots can be useful for model evaluation:

<b>Regression models:</b>
In regression tasks, where the goal is to predict continuous numerical values, you can plot the predicted values against the actual target values. A good model fit would show that the predicted values closely align with the actual values, resulting in a distribution plot that appears similar to a straight line (the 45-degree line if the model is perfectly accurate).

<b>Classification models: </b>
For classification tasks, where the goal is to predict categorical labels, you can create a distribution plot comparing the predicted class probabilities or scores to the true class labels. In an ideal scenario, the model's predicted probabilities for the correct class should be high, leading to a sharp peak around the correct class in the distribution plot.

<b>Residual plots:</b> 
When assessing the fit of a model, particularly in regression tasks, you can also plot the distribution of the model's residuals. Residuals are the differences between the predicted values and the actual values. A well-fitted model would have residuals distributed around zero with no apparent patterns or trends.

<b>Overfitting/Underfitting detection:</b>
Distribution plots can also help identify overfitting or underfitting. In overfitting, the model is too complex and captures noise in the training data, leading to poor generalization. This might be evident in a distribution plot as the predicted values deviating excessively from the actual values. <u>Underfitting</u>, on the other hand, occurs when the model is too simple to capture the underlying patterns in the data, resulting in a distribution plot that shows poor alignment between predicted and actual values.


In general, comparing the distributions of predicted and actual values can provide valuable insights into the model's performance and its ability to generalize to new, unseen data. However, it's essential to use a combination of evaluation metrics and visualization techniques to obtain a comprehensive assessment of the model's fit. Additionally, cross-validation or hold-out testing can further validate the model's performance on unseen data.

# 3- Polynomial Regression and Pipelines

Earlier, when we used "highway-mpg" as the predictor variable and fitted a linear model, it became evident that the model did not provide the best fit to the data. To explore alternative approaches, we will now consider fitting a polynomial model to the data. Polynomial models can capture more complex relationships between the predictor and the target variable, which might result in a better fit. Let's proceed with this approach to see if it leads to improved results.

```python
#plot the polynomial curve and visualize the fit of the model.
#independent variable as input and returns the predicted values of the dependent variable.
#<independent variable> polynomial model will make predictions. It's used as the x-coordinates for generating the new data points.
#<dependent variable> It's used as the y-coordinates for generating the new data points. This is the output data (response) used during the model's training phase. 
# Name:is a string that represents title of the plot 


def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of Cars')

   # plt.legend(loc="upper right", labels=["Price", "mpg"])
     
    plt.show()
    plt.close()
```

<b>the dependent_variable</b> is not used for making predictions in the given function. The dependent_variable is the actual output data (response) used during the model's training phase. It is not used for prediction but rather for comparing the model's predictions against the true values during the training process to adjust the model's parameters (coefficients) to minimize the prediction errors.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

The final step in the pipeline is typically an estimator, which is the machine learning model you want to train and use for making predictions. 

This could be any supervised learning algorithm such as a linear regression, support vector machine, or a deep learning model.

We create the pipeline by creating a <b>list of tuples</b> including the name of the model

```python
Input=[('scale',StandardScaler()), ('polynomial', PolynomialFeatures(include_bias=False)), ('model',LinearRegression())]
```

>Let's break down each step:

<b>Step 1: Data Scaling (StandardScaler)</b>

    ('scale', StandardScaler())
    
In this step, the StandardScaler class from scikit-learn is used to standardize the features of the dataset. Standardization scales the features in such a way that they have zero mean and unit variance, which can be important for certain algorithms (like linear regression) to work effectively, as it avoids any undue influence of features with larger scales.

<b>Step 2: Polynomial Feature Transformation (PolynomialFeatures)</b>
    
    ('polynomial', PolynomialFeatures(include_bias=False))
    

In this step, the PolynomialFeatures class from scikit-learn is used to perform polynomial feature transformation. Polynomial features are additional features created by raising the original features to a certain power. For example, if you have a feature x, the polynomial transformation of degree 2 will create additional features x^2, x^3, and so on. The include_bias=False argument ensures that a bias column (intercept term) is not added to the dataset

<b>Step 3: Linear Regression Modeling (LinearRegression)</b>

    ('model', LinearRegression())

In this step, the LinearRegression class from scikit-learn is used to train a linear regression model on the dataset. Linear regression is a supervised learning algorithm used for predicting a continuous target variable based on one or more input features. After the feature scaling and polynomial feature transformation (if applied), the linear regression model is trained to find the best-fit line that minimizes the error between the predicted values and the actual target values.

>R-squared

statistical measure used to evaluate how well a regression model fits the observed data.The R-squared value is a number between 0 and 1, If the R-squared value is close to 1, it indicates that the model explains a large portion of the variance, and it is a good fit for the data. 

R-squared = 1 - (Sum of squared residuals / Total sum of squares)

R-squared = 1 - (MSE of regression line / MSE of the avreage of data)

> Mean Squared Error (MSE)

<p>The Mean Squared Error measures the average of the squares of errors. That is, the difference between actual value (y) and the estimated value (ŷ).</p>
